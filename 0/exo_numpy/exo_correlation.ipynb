### Veille sur la Coor√©lation
Qu'est ce ?
    The statistical relationship between two variables is referred to as their correlation.

A quoi cela sert-til ?
    It can be useful in data analysis and modeling to better understand the relationships between variables.

Quand l'utilise ton ?
    We may also be interested in the correlation between input variables with the output variable in order provide insight into which variables may or may not be relevant as input for developing a model.

    The performance of some algorithms can deteriorate if two or more variables are tightly related, called multicollinearity. An example is linear regression, where one of the offending correlated variables should be removed in order to improve the skill of the model.

Comment l'utilise ton ? (lib/method?)
    scipy.stats.pearsonr
    numpy.random.cov


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
def print_correlation(set01, set02):
    print('scipy.stats.pearsonr:\n', pearsonr(set01,set02)[0])
    print('np.corrcoef:\n', np.corrcoef(set01,set02))
```

```python
set01 = np.random.rand(100)
set02 = np.random.rand(100)
set03 = np.random.rand(100)

np.corrcoef([set01,set02,set03])
# print_correlation(set01, set02)
```

```python
# np.random.seed(42)

set01 = 20 * np.random.randn(100) + 100
set02 = set01 + (10 + np.random.randn(100) + 50)

print_correlation(set01, set02)
```

```python
set01 = np.random.normal(30,10,size=1000)
set02 = np.random.normal(30,60,size=1000)

print_correlation(set01,set02)

plt.hist(set01,label='set01')
plt.hist(set02,alpha=0.4,label='set02')
plt.legend()
plt.show()
```

```python
gaus = np.linspace(0,10,500)
sin = np.sin(gaus)
cos = np.cos(gaus)

print_correlation(sin, cos)
plt.plot(gaus,sin,'-r')
plt.plot(gaus,cos,'-b')
```
