```python
import numpy as np
import matplotlib.pyplot as plt
```

Tracer un graphe linéaire des points (x, y) où x = [0, 1, 2, 3, 4] et y = [1, 4, 3, 8, 5].

```python
x = np.array([0,1,2,3,4])
y = np.array([1,4,3,8,8])

plt.scatter(x,y, color='blue')
plt.plot(x,y, color='pink')
plt.title('Graphe linéaire', loc='left')
plt.xlabel('De 0 à 4',loc='left')
plt.ylabel('De 1 à 8',loc='bottom')
plt.grid()
plt.show()
```

Tracer un histogramme à partir de données aléatoires générées par NumPy.

```python
plt.hist(np.random.randn(100000)*10,500,density=True,cumulative=1)
#plt.axis([100,400,500,80])
plt.text(-40,.04,r'$\mu=100,\ \sigma=15$')
plt.grid(axis='x')
plt.show()
```

Tracer un diagramme à barres à partir des données suivantes : 
categories = ['A', 'B', 'C', 'D', 'E'], valeurs = [25, 30, 15, 20, 10]

```python
categories = ['A', 'B', 'C', 'D', 'E']
valeurs = [25, 30, 15, 20, 10]

plt.bar(categories,valeurs,width=0.5,color='red',edgecolor='blue')
plt.grid(axis='y')
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
valeurs = [25, 30, 15, 20, 10]

plt.barh(categories,valeurs,color='green',edgecolor='blue')
plt.grid(axis='x')
```

Tracer un diagramme en barres empilées à partir des données suivantes :
categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25], valeurs2 = [5, 10, 15, 20]

```python
categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25]
valeurs2 = [5, 10, 15, 20]

fig, ax = plt.subplots()

bar = ax.bar(categories,valeurs1,label='Valeur 1')
bar = ax.bar(categories,valeurs2,label='Valeur 2')
ax.bar_label(bar, label_type='center')
ax.legend()
plt.show()
```

Tracer deux courbes sur le même graphe à partir des données suivantes :
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fmt1 = '>--g'
fmt2 = '<--r'
#fmt = '[marker][line][color]
plt.plot(x,y1,fmt1,label='sin')
plt.plot(x,y2,fmt2,label='cos')
plt.xlabel('linspace(0, 10, 100)')
plt.legend()
plt.grid()
plt.show()
```

Tracer une courbe polynomiale à partir des données suivantes :
x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

```python
x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

plt.plot(x,y, label='f(x) = 3x³ + 2x² - 8x +5',color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.axvline()
plt.axhline()
plt.show()
```
